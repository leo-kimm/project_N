```python
import pandas as pd
import glob

files=glob.glob('삼숭*.txt')

samsung = []
for filename in files:
    samsung.append(pd.read_csv(filename, sep='\t', encoding='cp949'))

# Concatenate all data into one DataFrame
samsung = pd.concat(samsung, ignore_index=True)
```

```python
samsung
```

```python
samsung = samsung.sort_values(['일자'])
samsung
```

```python
samsung.to_csv('samsung10', index = False)
```
